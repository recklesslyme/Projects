- Given all the metrics in the final merged dataframe, it is difficult to make comparisons currently.
- It would be better to make certain features that are combinations of columns to help better understand all the metrics per country.
- Population, Population Density & Household exp per Capita, will be left as is, and factored separately. 
- That leaves 6 index factors from; Rent Index to Property Price to Income Ratio. 
- These 6 factors can be broadly categorized into two:
1. **Negative Indices** : Those metrics that are worse the higher the score is. For example; Cost of living + Rent index - which is the overall cost of healthcare, groceries, living expenses & rent. The higher it is in countries, the higher expenditure the population makes in general for these criteria. The Negative Indices in this set are; Cost of Living Plus Rent Index & Property Price to Income Ratio. (Cost of living plus Rent index = Rent Index + cost of living)
2. **Positive Indices** : Those metrics that are better the higher the score is. Here that would consist of: Purchasing Power Index & Quality of Life Index
3. All population & age structure related metrics will be evaluated in combination separately
4. Population & Pop Density will be retained only as new features
5. Age 15 to 64 years % will be retained - Since that is the primary target demographic
6. New features for Population is ; Population between 15 to 64 years & for Pop Density/mi2 is

**scaling the data is required here, since all the indices aren't on the same scale, comparing them could show lopsided or meaningless results**

```python
# Scaling all Index data

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

indices_list =['Cost of Living Plus Rent Index',
       'Quality of Life Index', 'Purchasing Power Index', 
               'Property Price to Income Ratio']

indices = df4.loc[:,indices_list]

indices.sample(5)
```

```python
# fitting the scaler transformation
df5 = sc.fit_transform(indices)

# creating a df out of the result
df5 = pd.DataFrame(df5)

# Rename column names as per indices.sample code

df5=df5.rename(columns={0:'Cost of Living Plus Rent Index',
                    1:'Quality of Life Index',
                    2:'Purchasing Power Index',
                    3:'Property Price to Income Ratio'})

df5.sample(5)
```

```python
# Creation of 3 new features

# Negative Indices: Cost of liv + rent index + Property Price to income ratio
# The lowest scores will already be in negative, therefore the smallest of
# Negatives will have the least cost of living & property price ratio

df5['Negatives'] = df5['Cost of Living Plus Rent Index'] + df5['Property Price to Income Ratio']



# positive Indices: Quality of live + purchasing power
df5['Positives'] = df5['Purchasing Power Index'] + df5['Quality of Life Index']

# Combined Indices by simply subtracting the two
# Since best total values are wanted; this implies selecting Positives with
# high positive scores & negatives with the lowest scores, i.e; 
# highest scores in the negative direction
# Therefore the countries with best Total scores would be those
# with the highest totals since: we compute High positive value - (high negative value)

df5['Total'] = df5['Positives'] - df5['Negatives']

# finally adding the country column from before

df5['Country'] = df4['Country']

# check

# reposition columns
df5 = df5[['Country','Total','Positives','Negatives','Cost of Living Plus Rent Index',
         'Quality of Life Index','Purchasing Power Index','Property Price to Income Ratio']]

df5.sample(5).round(2)
```

- merged dataset is ready to go. All plots and further computations in problem solving section
